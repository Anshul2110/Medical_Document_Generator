```python
import os
```

```python
from openai import AzureOpenAI



client = AzureOpenAI(
    api_key="xxxx",
    azure_endpoint="xxxxxx",
    api_version="xxx"
)
```

```python
prompt = f"""
    You are a professional medical report generator.

    IMPORTANT:
    - Do NOT hallucinate data
    - Only use provided information
    - This is NOT a final clinical diagnosis
    - Maintain formal tone

    Patient ID: {"pt_123"}
    Age: {21}
    Gender: {"Male"}

    Symptoms:
    {"Chest pain and shortness of breath"}

    X-ray Observations:
    {"Mild opacity in left lung"}

    Diagnosis:
    {"Possible early pneumonia"}

    Generate structured report with:
    1. Patient Summary
    2. Clinical Findings
    3. Diagnosis Explanation
    4. Recommendations
    5. Disclaimer
    """
```

```python
response = client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2,
    )
```

```python
result = response.choices[0].message.content
```

```python
print(result)
```

```python
from docx import Document
```

```python
doc = Document()

doc.add_heading("Medical Report", 0)
doc.add_paragraph(f"Patient ID: {"pt_123"}")

for line in result.split("\n"):
    doc.add_paragraph(line)

file_path = r"trial_docs\pt123.docx"
doc.save(file_path)
```
