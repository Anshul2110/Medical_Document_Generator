```python

from os import name

from cryptography.fernet import Fernet
import base64
# Generate a key (store this securely!)
key = Fernet.generate_key()
cipher = Fernet(key)

# def encrypt_name(name: str) -> str:
encrypted_name = cipher.encrypt("Anshul Desai".encode()).decode()

# def decrypt_name(encrypted_name: str) -> str:
decrypt_name = cipher.decrypt(encrypted_name.encode()).decode()
```
